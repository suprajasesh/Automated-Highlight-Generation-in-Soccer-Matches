```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
from tqdm.notebook import tqdm
from google.colab import drive
import gc
from torch.nn.utils.rnn import pad_sequence

# Mount Google Drive
drive.mount('/content/drive')

# Path to your SoccerNet dataset shortcut
SOCCERNET_PATH = "/content/drive/MyDrive/soccernet"

# Define constants
FPS = 2  # ResNet features are extracted at 2 frames per second
WINDOW_SIZE = 40  # 20-second window (20s * 2fps = 40 frames)
STEP_SIZE = 4    # Smaller step size for more dense predictions
EVENT_WINDOW = 4  # Number of frames around event to consider as event (+/-)
MAX_MATCHES = 100  # Limit to 100 matches

class EventMapper:
    """Maps event labels to numerical indices and vice versa"""

    def __init__(self):
        # List of event labels in SoccerNet-v2
        self.events = [
            "Ball out of play", "Throw-in", "Foul", "Indirect free-kick",
            "Clearance", "Shot", "Shot on target", "Goal", "Corner", "Substitution",
            "Kick-off", "Yellow card", "Offside", "Direct free-kick", "Red card",
            "Yellow->red card", "Penalty"
        ]
        # Add "Background" class for frames without events
        self.events.append("Background")

        # Create mappings
        self.event_to_idx = {event: i for i, event in enumerate(self.events)}
        self.idx_to_event = {i: event for i, event in enumerate(self.events)}

    def get_num_classes(self):
        return len(self.events)

    def event_to_index(self, event):
        return self.event_to_idx.get(event, self.event_to_idx["Background"])

    def index_to_event(self, idx):
        return self.idx_to_event[idx]

def game_time_to_frame(game_time, fps=FPS):
    """Convert game time string (e.g. '1 - 00:00') to frame number"""
    half, time_str = game_time.split(" - ")
    minutes, seconds = time_str.split(":")

    # Convert to seconds
    total_seconds = int(half) * 45 * 60 + int(minutes) * 60 + int(seconds)

    # Convert to frame number
    frame_number = total_seconds * fps

    return frame_number

def load_labels(json_path):
    """Load labels from SoccerNet json file and convert to frame-level annotations"""
    with open(json_path, 'r') as f:
        data = json.load(f)

    annotations = data.get('annotations', [])
    frame_labels = []

    for annotation in annotations:
        game_time = annotation['gameTime']
        label = annotation['label']
        position = int(annotation['position'])  # Frame position in the video
        team = annotation['team']
        visibility = annotation['visibility']

        # Calculate the frame number from game time
        frame_number = game_time_to_frame(game_time)

        # Only include visible events
        if visibility == "visible":
            frame_labels.append({
                'frame': frame_number,
                'label': label,
                'position': position,
                'team': team
            })

    return frame_labels

def get_match_paths():
    """Return all match paths from the SoccerNet dataset, limited to MAX_MATCHES"""
    match_paths = []

    # List of leagues
    leagues = ['england_epl', 'europe_uefa-champions-league', 'france_ligue-1',
               'germany_bundesliga', 'italy_serie-a', 'spain_laliga']

    # List of seasons
    seasons = ['2014-2015', '2015-2016', '2016-2017']

    # Traverse through all leagues and seasons
    for league in leagues:
        league_path = os.path.join(SOCCERNET_PATH, league)
        if not os.path.exists(league_path):
            print(f"Warning: League path {league_path} does not exist")
            continue

        for season in seasons:
            season_path = os.path.join(league_path, season)
            if not os.path.exists(season_path):
                print(f"Warning: Season path {season_path} does not exist")
                continue

            # Get all match directories
            for match_dir in os.listdir(season_path):
                match_path = os.path.join(season_path, match_dir)
                if os.path.isdir(match_path):
                    # Check if the required files exist
                    labels_path = os.path.join(match_path, 'Labels-v2.json')
                    half1_path = os.path.join(match_path, '1_ResNET_TF2.npy')
                    half2_path = os.path.join(match_path, '2_ResNET_TF2.npy')

                    if os.path.exists(labels_path) and os.path.exists(half1_path) and os.path.exists(half2_path):
                        match_paths.append(match_path)
                        if len(match_paths) >= MAX_MATCHES:  # Stop when we have enough matches
                            return match_paths

    return match_paths

class SoccerNetDataset(Dataset):
    """Memory-efficient SoccerNet Dataset with balanced sampling"""

    def __init__(self, match_paths, event_mapper, mode='train', feature_dim=2048):
        self.match_paths = match_paths
        self.event_mapper = event_mapper
        self.mode = mode
        self.feature_dim = feature_dim
        self.samples = []
        self.event_samples = {i: [] for i in range(event_mapper.get_num_classes())}

        # Cache to store loaded features to avoid reloading
        self.feature_cache = {}

        # Pre-compute the indices (but don't load features yet)
        self._precompute_indices()

        # Balance dataset if in training mode
        if mode == 'train':
            self._balance_dataset()
        else:
            # For validation/test, limit background samples
            self._limit_background_samples()

    def _precompute_indices(self):
        """Precompute all sample indices without loading features"""
        for match_idx, match_path in enumerate(tqdm(self.match_paths, desc=f"Indexing {self.mode} matches")):
            try:
                # Load label data
                labels_path = os.path.join(match_path, 'Labels-v2.json')
                frame_labels = load_labels(labels_path)

                # Create a mapping from frame number to event label for each frame
                frame_to_label = {}
                for event in frame_labels:
                    frame = event['frame']
                    label = event['label']

                    # Label the event frame and surrounding frames
                    for offset in range(-EVENT_WINDOW, EVENT_WINDOW + 1):
                        target_frame = frame + offset
                        if target_frame >= 0:  # Ensure non-negative frame indices
                            frame_to_label[target_frame] = label

                # Process each half
                for half_idx in [1, 2]:
                    # Get the expected number of frames without loading features
                    half_file = os.path.join(match_path, f'{half_idx}_ResNET_TF2.npy')
                    if not os.path.exists(half_file):
                        continue

                    # Get array shape without loading whole array
                    with open(half_file, 'rb') as f:
                        version = np.lib.format.read_magic(f)
                        shape, _, _ = np.lib.format.read_array_header_1_0(f)
                        num_frames = shape[0]

                    # Create sliding window indices
                    for i in range(0, num_frames - WINDOW_SIZE + 1, STEP_SIZE):
                        middle_frame = i + WINDOW_SIZE // 2

                        # Adjust frame number to match global frame number
                        global_middle_frame = middle_frame
                        if half_idx == 2:
                            global_middle_frame += 45 * 60 * FPS  # Add 45 minutes worth of frames

                        # Determine the label
                        label = frame_to_label.get(global_middle_frame, "Background")
                        label_idx = self.event_mapper.event_to_index(label)

                        # Skip some background samples in training mode
                        if label == "Background" and self.mode == 'train':
                            # In training mode, keep only some background samples
                            if np.random.random() > 0.03:  # Keep only 3% of background samples
                                continue

                        sample_info = (match_path, half_idx, i, middle_frame, label_idx)

                        # Add to samples list
                        self.samples.append(sample_info)

                        # Also add to event-specific list for balancing
                        self.event_samples[label_idx].append(sample_info)

            except Exception as e:
                print(f"Error processing {match_path}: {str(e)}")
                continue

    def _balance_dataset(self):
        """Balance the dataset by oversampling minority classes"""
        # Count samples per class
        class_counts = {i: len(samples) for i, samples in self.event_samples.items()}
        print(f"Original class distribution: {class_counts}")

        # Find the class with maximum samples (excluding background)
        background_idx = self.event_mapper.event_to_index("Background")
        event_counts = {cls: count for cls, count in class_counts.items() if cls != background_idx}

        if not event_counts:
            print("Warning: No event samples found!")
            return

        # Increase target count for better representation of minority classes
        target_count = max(100, max(event_counts.values()))  # At least 100 samples per class

        # Limit background samples to avoid extreme imbalance
        background_limit = min(target_count, len(self.event_samples[background_idx]))

        # Create balanced samples list
        balanced_samples = []

        # Add non-background samples with oversampling
        for cls, samples in self.event_samples.items():
            if len(samples) == 0:
                # Skip empty classes
                continue

            if cls == background_idx:
                # For background, limit the number
                if len(samples) > background_limit:
                    # Convert samples to array for choice function
                    sampled_indices = np.random.choice(len(samples), background_limit, replace=False)
                    balanced_samples.extend([samples[i] for i in sampled_indices])
                else:
                    balanced_samples.extend(samples)
            else:
                # For event classes, oversample to reach target_count
                # Higher oversampling for rare events
                if len(samples) < 20:  # Very rare events
                    oversample_factor = 5
                else:
                    oversample_factor = 3

                target = min(len(samples) * oversample_factor, target_count)
                sampled_indices = np.random.choice(len(samples), target, replace=len(samples) < target)
                balanced_samples.extend([samples[i] for i in sampled_indices])

        # Update samples list
        if len(balanced_samples) > 0:
            self.samples = balanced_samples

            # Recalculate class distribution
            new_distribution = {}
            for sample in self.samples:
                label = sample[4]
                if label not in new_distribution:
                    new_distribution[label] = 0
                new_distribution[label] += 1

            print(f"Balanced dataset: {len(self.samples)} samples")
            print(f"New class distribution: {new_distribution}")

    def _limit_background_samples(self):
        """Limit background samples in validation/test set"""
        background_idx = self.event_mapper.event_to_index("Background")
        background_samples = self.event_samples[background_idx]

        # Count total event samples
        total_event_samples = sum(len(samples) for cls, samples in self.event_samples.items()
                                if cls != background_idx)

        # Limit background samples to at most 2x the number of event samples
        background_limit = min(len(background_samples), total_event_samples * 2)

        if len(background_samples) > background_limit:
            # Sample a subset of background samples
            sampled_indices = np.random.choice(len(background_samples), background_limit, replace=False)
            sampled_background = [background_samples[i] for i in sampled_indices]

            # Create new samples list with all event samples and limited background
            new_samples = []
            for sample in self.samples:
                if sample[4] != background_idx:
                    new_samples.append(sample)

            new_samples.extend(sampled_background)
            self.samples = new_samples

            print(f"Limited {self.mode} background samples: {len(background_samples)} -> {background_limit}")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        match_path, half_idx, window_start, middle_frame, label = self.samples[idx]

        try:
            # Use cache key to avoid reloading the same files
            cache_key = (match_path, half_idx)

            if cache_key in self.feature_cache:
                features = self.feature_cache[cache_key]
            else:
                # Load features fully into memory (no memory mapping)
                half_file = os.path.join(match_path, f'{half_idx}_ResNET_TF2.npy')
                features = np.load(half_file)

                # Store in cache if we're not in training mode (to save memory)
                if self.mode != 'train':
                    self.feature_cache[cache_key] = features

            # Check valid range
            if window_start + WINDOW_SIZE > len(features):
                # Handle edge case where window extends beyond available frames
                available_frames = len(features) - window_start
                window_features = np.zeros((WINDOW_SIZE, self.feature_dim))
                window_features[:available_frames] = features[window_start:window_start+available_frames]
            else:
                # Get the window features
                window_features = features[window_start:window_start+WINDOW_SIZE]

            # Apply better feature normalization
            window_features = (window_features - np.mean(window_features, axis=0, keepdims=True)) / \
                             (np.std(window_features, axis=0, keepdims=True) + 1e-5)

            # Add temporal difference features (helps detect motion patterns)
            if window_features.shape[0] > 1:
                diff_features = np.diff(window_features, axis=0)
                diff_features = np.vstack([np.zeros((1, self.feature_dim)), diff_features])  # Padding for first frame

                # Combine original and difference features
                combined_features = np.concatenate([window_features, diff_features], axis=1)
            else:
                # Handle edge case with a single frame
                combined_features = np.concatenate([window_features, np.zeros_like(window_features)], axis=1)

            return torch.tensor(combined_features, dtype=torch.float32), torch.tensor(label, dtype=torch.long)

        except Exception as e:
            print(f"Error loading sample {idx}: {str(e)}")
            # Return a dummy sample
            dummy_features = torch.zeros((WINDOW_SIZE, self.feature_dim * 2), dtype=torch.float32)  # *2 for diff features
            dummy_label = torch.tensor(self.event_mapper.event_to_index("Background"), dtype=torch.long)
            return dummy_features, dummy_label

    def cleanup_cache(self):
        """Clear feature cache to free memory"""
        self.feature_cache.clear()
        gc.collect()

class EventDetectionModel(nn.Module):
    """Advanced neural network for soccer event detection"""

    def __init__(self, input_dim, hidden_dim, num_classes, dropout_rate=0.5):
        super(EventDetectionModel, self).__init__()

        # Input dimension reduction - input now includes difference features
        self.input_dim = input_dim * 2  # Original features + temporal difference features

        # Feature reduction
        self.feature_reducer = nn.Sequential(
            nn.Linear(self.input_dim, hidden_dim * 2),
            nn.LeakyReLU(0.1),
            nn.BatchNorm1d(WINDOW_SIZE),
            nn.Dropout(dropout_rate)
        )

        # 1D convolutional layers with different kernel sizes for multi-scale temporal features
        self.conv1 = nn.Conv1d(hidden_dim * 2, hidden_dim, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(hidden_dim * 2, hidden_dim, kernel_size=5, padding=2)
        self.conv3 = nn.Conv1d(hidden_dim * 2, hidden_dim, kernel_size=7, padding=3)

        # Bi-directional LSTM layers
        self.lstm1 = nn.LSTM(
            input_size=hidden_dim * 3,  # Combined conv outputs
            hidden_size=hidden_dim,
            num_layers=2,
            batch_first=True,
            bidirectional=True,
            dropout=dropout_rate
        )

        # Attention mechanism
        self.attention = nn.Sequential(
            nn.Linear(hidden_dim * 2, 64),
            nn.Tanh(),
            nn.Linear(64, 1)
        )

        # Output layers with skip connection
        self.fc1 = nn.Linear(hidden_dim * 2, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.classifier = nn.Linear(hidden_dim // 2, num_classes)

        # Additional skip connection
        self.skip_connection = nn.Linear(hidden_dim * 2, hidden_dim // 2)

        # Batch normalization layers
        self.bn1 = nn.BatchNorm1d(hidden_dim * 3)
        self.bn2 = nn.BatchNorm1d(hidden_dim * 2)
        self.bn3 = nn.BatchNorm1d(hidden_dim)
        self.bn4 = nn.BatchNorm1d(hidden_dim // 2)

    def forward(self, x):
        batch_size, seq_len, feat_dim = x.size()

        # Apply feature reduction
        orig_x = x.clone()  # Save for residual connection
        x = x.view(-1, feat_dim)
        x = self.feature_reducer(x.view(batch_size, seq_len, -1))

        # Multi-scale temporal convolution
        x_perm = x.permute(0, 2, 1)  # [batch, hidden_dim*2, seq_len]

        conv1_out = torch.relu(self.conv1(x_perm))
        conv2_out = torch.relu(self.conv2(x_perm))
        conv3_out = torch.relu(self.conv3(x_perm))

        # Concatenate convolutional outputs
        conv_combined = torch.cat([conv1_out, conv2_out, conv3_out], dim=1)

        # Apply batch normalization
        conv_combined = self.bn1(conv_combined)

        # Pass through LSTM
        lstm_in = conv_combined.permute(0, 2, 1)  # [batch, seq_len, hidden_dim*3]
        lstm_out, _ = self.lstm1(lstm_in)  # [batch, seq_len, hidden_dim*2]

        # Apply batch normalization
        lstm_bn = self.bn2(lstm_out.permute(0, 2, 1)).permute(0, 2, 1)

        # Apply attention mechanism
        attn_weights = self.attention(lstm_bn).squeeze(-1)  # [batch, seq_len]
        attn_weights = torch.softmax(attn_weights, dim=1).unsqueeze(-1)  # [batch, seq_len, 1]

        # Context vector is weighted sum of LSTM outputs
        context = torch.sum(lstm_bn * attn_weights, dim=1)  # [batch, hidden_dim*2]

        # Feed-forward layers with skip connection
        out1 = torch.relu(self.fc1(context))
        out1 = self.bn3(out1)
        out1 = nn.functional.dropout(out1, p=0.4, training=self.training)

        out2 = torch.relu(self.fc2(out1))
        out2 = self.bn4(out2)

        # Skip connection
        skip = self.skip_connection(context)
        out = out2 + skip

        # Final classification
        logits = self.classifier(out)

        return logits

def focal_loss(outputs, targets, alpha=0.5, gamma=2.0):
    """Focal loss for addressing class imbalance"""
    ce_loss = torch.nn.functional.cross_entropy(outputs, targets, reduction='none')
    pt = torch.exp(-ce_loss)
    loss = alpha * (1-pt)**gamma * ce_loss
    return loss.mean()

def train_model(model, train_loader, val_loader, optimizer, num_epochs=3, device='cuda'):
    """Train the model with early stopping and learning rate scheduling"""
    # Move model to device
    model.to(device)

    # Training history
    train_losses = []
    val_losses = []
    val_accuracies = []

    # Early stopping variables
    best_val_loss = float('inf')
    best_model_state = None
    patience = 5  # Increased patience
    patience_counter = 0

    # Learning rate scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, 'min', patience=2, factor=0.5, verbose=True
    )

    for epoch in range(num_epochs):
        # Training
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0
        background_idx = val_loader.dataset.event_mapper.event_to_index("Background")

        train_preds = []
        train_labels = []

        for batch_idx, (features, labels) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Training")):
            features = features.to(device)
            labels = labels.to(device)

            # Zero the parameter gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(features)

            # Calculate loss using focal loss for better handling of imbalanced classes
            loss = focal_loss(outputs, labels, alpha=0.75, gamma=2.0)

            # Backward pass and optimize
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()

            running_loss += loss.item() * features.size(0)

            # Calculate accuracy
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Track predictions and labels for metrics
            train_preds.extend(predicted.cpu().numpy())
            train_labels.extend(labels.cpu().numpy())

            # Print batch statistics periodically
            if batch_idx % 20 == 0:
                batch_acc = (predicted == labels).sum().item() / labels.size(0)
                print(f"  Batch {batch_idx}: Loss={loss.item():.4f}, Acc={batch_acc:.4f}")

        epoch_train_loss = running_loss / len(train_loader.dataset)
        epoch_train_acc = correct / total
        train_losses.append(epoch_train_loss)

        # Print training metrics excluding background class
        non_bg_correct = 0
        non_bg_total = 0
        for i in range(len(train_labels)):
            if train_labels[i] != background_idx:
                non_bg_total += 1
                if train_labels[i] == train_preds[i]:
                    non_bg_correct += 1

        non_bg_acc = non_bg_correct / max(1, non_bg_total)
        print(f"Training - Event classes only: {non_bg_acc:.4f} ({non_bg_correct}/{non_bg_total})")

        # Validation
        val_metrics = evaluate_during_training(model, val_loader, device)
        epoch_val_loss = val_metrics['loss']
        epoch_val_acc = val_metrics['accuracy']
        epoch_val_event_acc = val_metrics['event_accuracy']

        val_losses.append(epoch_val_loss)
        val_accuracies.append(epoch_val_acc)

        print(f"Epoch {epoch+1}/{num_epochs} - "
              f"Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f}, "
              f"Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}, "
              f"Val Event Acc: {epoch_val_event_acc:.4f}")

        # Print F1 scores for validation
        print(f"Validation F1 Scores: {val_metrics['f1_scores']}")

        # Check for early stopping
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            best_model_state = model.state_dict().copy()
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping triggered after {epoch+1} epochs")
                break

        # Update learning rate
        scheduler.step(epoch_val_loss)

        # Clean up memory
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    # Load the best model
    if best_model_state is not None:
        model.load_state_dict(best_model_state)

    # Additional metrics
    metrics = {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'val_accuracies': val_accuracies
    }

    return model, metrics

def evaluate_during_training(model, data_loader, device):
    """Evaluate model during training"""
    model.eval()
    total_loss = 0
    correct = 0
    total = 0

    all_preds = []
    all_labels = []

    background_idx = data_loader.dataset.event_mapper.event_to_index("Background")

    with torch.no_grad():
        for features, labels in data_loader:
            features = features.to(device)
            labels = labels.to(device)

            outputs = model(features)
            loss = focal_loss(outputs, labels)

            total_loss += loss.item() * features.size(0)

            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculate event-only accuracy (excluding background)
    non_bg_correct = 0
    non_bg_total = 0
    for i in range(len(all_labels)):
        if all_labels[i] != background_idx:
            non_bg_total += 1
            if all_labels[i] == all_preds[i]:
                non_bg_correct += 1

    # Calculate class-specific F1 scores
    from sklearn.metrics import f1_score

    # Convert to numpy arrays
    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)

    # Calculate micro and macro F1 scores
    micro_f1 = f1_score(all_labels, all_preds, average='micro')
    macro_f1 = f1_score(all_labels, all_preds, average='macro')

    event_accuracy = non_bg_correct / max(1, non_bg_total)

    metrics = {
        'loss': total_loss / len(data_loader.dataset),
        'accuracy': correct / total,
        'event_accuracy': event_accuracy,
        'f1_scores': {
            'micro': micro_f1,
            'macro': macro_f1
        }
    }

    return metrics

def evaluate_model(model, test_loader, event_mapper, device='cuda'):
    """Evaluate the trained model"""
    model.eval()

    all_predictions = []
    all_labels = []
    background_idx = event_mapper.event_to_index("Background")

    with torch.no_grad():
        for features, labels in tqdm(test_loader, desc="Evaluating"):
            features = features.to(device)
            outputs = model(features)
            _, predictions = torch.max(outputs, 1)

            all_predictions.extend(predictions.cpu().numpy())
            all_labels.extend(labels.numpy())

    # Convert to numpy arrays
    all_predictions = np.array(all_predictions)
    all_labels = np.array(all_labels)

    # Extract non-background samples for separate evaluation
    event_mask = all_labels != background_idx
    event_predictions = all_predictions[event_mask]
    event_labels = all_labels[event_mask]

    # Find unique classes in the test data
    unique_classes = np.unique(np.concatenate([all_labels, all_predictions]))
    target_names = [event_mapper.index_to_event(i) for i in unique_classes]

    # Calculate metrics
    print("Overall Results:")
    overall_report = classification_report(
        all_labels,
        all_predictions,
        labels=unique_classes,
        target_names=target_names,
        zero_division=0
    )
    print(overall_report)

    if len(event_labels) > 0:
        print("\nEvent-Only Results (excluding background):")
        event_unique_classes = np.unique(np.concatenate([event_labels, event_predictions]))
        event_target_names = [event_mapper.index_to_event(i) for i in event_unique_classes]

        event_report = classification_report(
            event_labels,
            event_predictions,
            labels=event_unique_classes,
            target_names=event_target_names,
            zero_division=0
        )
        print(event_report)

    # Create confusion matrix
    plt.figure(figsize=(15, 15))
    cm = confusion_matrix(all_labels, all_predictions, labels=unique_classes)

    # Normalize confusion matrix
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    cm_normalized = np.nan_to_num(cm_normalized)  # Replace NaN with 0

    sns.heatmap(
        cm_normalized,
        annot=True,
        fmt='.2f',
        cmap='Blues',
        xticklabels=target_names,
        yticklabels=target_names
    )
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Normalized Confusion Matrix')
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    return overall_report, cm

def predict_events(model, match_path, event_mapper, device='cuda'):
    """Predict events for a given match"""
    model.eval()

    # Load ResNet features
    half1_features = np.load(os.path.join(match_path, '1_ResNET_TF2.npy'))
    half2_features = np.load(os.path.join(match_path, '2_ResNET_TF2.npy'))

    predictions = []
    background_idx = event_mapper.event_to_index("Background")

    for half_idx, features in enumerate([half1_features, half2_features], 1):
        num_frames = len(features)
        half_predictions = np.full(num_frames, background_idx)

        # Create sliding windows
        for i in range(0, num_frames - WINDOW_SIZE + 1, STEP_SIZE):
            window_features = features[i:i+WINDOW_SIZE]

            # Normalize features
            window_features = (window_features - window_features.mean(axis=0)) / (window_features.std(axis=0) + 1e-5)

            # Add temporal difference features
            if window_features.shape[0] > 1:
                diff_features = np.diff(window_features, axis=0)
                diff_features = np.vstack([np.zeros((1, window_features.shape[1])), diff_features])
                combined_features = np.concatenate([window_features, diff_features], axis=1)
            else:
                combined_features = np.concatenate([window_features, np.zeros_like(window_features)], axis=1)

            # Convert to tensor
            window_tensor = torch.tensor(combined_features, dtype=torch.float32).unsqueeze(0).to(device)

            # Make prediction
            with torch.no_grad():
                outputs = model(window_tensor)
                _, predicted = torch.max(outputs, 1)
                predicted_class = predicted.item()

            # Only update if event is detected (not background)
            if predicted_class != background_idx:
                # Update prediction for the middle frame of the window
                middle_frame = i + WINDOW_SIZE // 2
                if middle_frame < num_frames:
                    half_predictions[middle_frame] = predicted_class

        # Add half predictions to the list
        predictions.append(half_predictions)

    # Convert predictions to event sequences
    events = []
    for half_idx, half_predictions in enumerate(predictions, 1):
        frame_idx = 0
        while frame_idx < len(half_predictions):
            class_idx = half_predictions[frame_idx]

            # Skip background frames
            if class_idx != background_idx:
                event_name = event_mapper.index_to_event(class_idx)

                # Calculate game time
                seconds = frame_idx / FPS
                minutes = int(seconds // 60)
                remaining_seconds = int(seconds % 60)

                # Add 45 minutes for second half
                if half_idx == 2:
                    minutes += 45

                game_time = f"{half_idx} - {minutes:02d}:{remaining_seconds:02d}"

                events.append({
                    'gameTime': game_time,
                    'label': event_name,
                    'half': half_idx,
                    'frame': frame_idx,
                    'confidence': 1.0  # We could add confidence scores here
                })

            frame_idx += 1

    return events, predictions

def main():
    # Check if GPU is available
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Create event mapper
    event_mapper = EventMapper()
    print(f"Number of event classes: {event_mapper.get_num_classes()}")

    # Get all match paths (limited to MAX_MATCHES)
    match_paths = get_match_paths()
    print(f"Found {len(match_paths)} matches (limited to first {MAX_MATCHES})")

    # Split matches into train, validation, and test sets (70/15/15 split)
    train_matches, temp_matches = train_test_split(match_paths, test_size=0.3, random_state=42)
    val_matches, test_matches = train_test_split(temp_matches, test_size=0.5, random_state=42)

    print(f"Training matches: {len(train_matches)}")
    print(f"Validation matches: {len(val_matches)}")
    print(f"Test matches: {len(test_matches)}")

    # Create datasets with balanced sampling
    print("Creating training dataset...")
    train_dataset = SoccerNetDataset(train_matches, event_mapper, mode='train')

    print("Creating validation dataset...")
    val_dataset = SoccerNetDataset(val_matches, event_mapper, mode='val')

    print("Creating test dataset...")
    test_dataset = SoccerNetDataset(test_matches, event_mapper, mode='test')

    # Create data loaders with smaller batch sizes
    batch_size = 32  # Reduced from 64 to save memory
    num_workers = 2

    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=num_workers,
        pin_memory=True  # Helps with GPU transfer
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=num_workers,
        pin_memory=True
    )

    test_loader = DataLoader(
        test_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=num_workers,
        pin_memory=True
    )

    # Create model
    input_dim = 2048  # ResNet features are 2048-dimensional
    hidden_dim = 256
    num_classes = event_mapper.get_num_classes()

    model = EventDetectionModel(input_dim, hidden_dim, num_classes, dropout_rate=0.5)
    print(model)

    # Define optimizer - We'll use focal loss directly in the training function
    optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4)

    # Train model
    num_epochs = 3
    model, metrics = train_model(
        model,
        train_loader,
        val_loader,
        optimizer,
        num_epochs=num_epochs,
        device=device
    )

    # Clean up memory after training
    train_dataset.cleanup_cache()
    val_dataset.cleanup_cache()
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    # Plot training curves
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(metrics['train_losses'], label='Train Loss')
    plt.plot(metrics['val_losses'], label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(metrics['val_accuracies'], label='Val Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/soccernet_training_curves.png')
    plt.show()

    # Evaluate model
    report, cm = evaluate_model(model, test_loader, event_mapper, device=device)

    # Save model
    torch.save({
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'train_losses': metrics['train_losses'],
        'val_losses': metrics['val_losses'],
        'val_accuracies': metrics['val_accuracies'],
    }, '/content/drive/MyDrive/soccernet_event_detection_model.pth')

    # Example: Predict events for a test match
    if len(test_matches) > 0:
        test_match = test_matches[0]
        print(f"Predicting events for match: {os.path.basename(test_match)}")

        events, predictions = predict_events(model, test_match, event_mapper, device=device)

        # Display predictions
        print(f"Predicted {len(events)} events:")
        for event in events[:10]:  # Display first 10 events
            print(f"{event['gameTime']}: {event['label']}")

        # Compare with ground truth
        labels_path = os.path.join(test_match, 'Labels-v2.json')
        if os.path.exists(labels_path):
            true_events = load_labels(labels_path)
            print(f"\nGround truth: {len(true_events)} events")
            for event in true_events[:10]:  # Display first 10 events
                print(f"{event['frame'] / FPS / 60:.2f} min: {event['label']}")

    # Clean up remaining memory
    test_dataset.cleanup_cache()
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    print("Done!")

if __name__ == "__main__":
    main()
```
